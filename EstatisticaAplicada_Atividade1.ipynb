```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import io

from google.colab import files


uploaded = files.upload()
```

Fazer a leitura de arquivo .csv e em seguida colocar ele em um dataframe:

```python
df = pd.read_csv(io.BytesIO(uploaded['tabela1.csv']), sep=";")
df
```

Anlises e Grficos:

```python
df_sexo = pd.crosstab(index=df['Turma'], columns=df['Sexo'])
df_sexo
```

A tabela acima faz a relao entre as turmas e quantidade de alunos por sexo.  possvel verificar que tanto na turma A quanto na turma o nmero de mulheres  maior que a de homens, isto pode ser um indicativo que o curso atrai mais o pblico feminino.

```python
# Filtrando apenas as linhas onde 'Exerc'  maior que 0
df_filtered = df[(df['Exerc'] > 0) & (df['Sexo'] != 'NI')]
df_filtered

# Usando a pd.crosstab para criar a tabela cruzada
crosstab_result = pd.crosstab(index=df_filtered['Sexo'], columns='Pessoas que fazem exerccios')
crosstab_result['Total de pessoas'] = df['Sexo'].value_counts()
crosstab_result ['Porcentagem (%)'] = (crosstab_result['Pessoas que fazem exerccios'] / (crosstab_result['Total de pessoas'])) * 100
crosstab_result = crosstab_result.rename_axis("", axis=1)

crosstab_result
```

Na tabela acima  feito a correlao entre sexo e a prtica de exerccio fsico, e tambm  gerado a proporo desta prtica em relao aos sexos. Com esses dados em mos nota-se que os homens praticam mais exerccio fsico do que as mulheres de forma proporcional.

```python
#Cria filtros de dataframe para auxiliar na construo da tabela
df_filtro_fuma_exerc = df[(df['Fuma'] == 'SIM') & (df['Exerc'] > 0)]
df_filtro_fuma_nexerc = df[(df['Fuma'] == 'SIM') & (df['Exerc'] == 0)]
#df_fuma = pd.crosstab(index=df[(df['Fuma'] == 'SIM') & (df['Exerc'] > 0)]['Sexo'], columns='FumantesExerccio')

# Criar uma tabela de frequncia cruzada
df_fuma = pd.crosstab(index=df_filtro_fuma_exerc['Sexo'], columns='FumantesExerccio')
df_fuma['FumantesSemExerccio'] = df_filtro_fuma_nexerc['Sexo'].value_counts()
df_fuma['Fuma'] = df['Fuma'].value_counts()

crosstab_result['Praticam exerccio e fumam'], crosstab_result['No praticam exerccio e fumam'] = df_fuma['FumantesExerccio'], df_fuma['FumantesSemExerccio'].fillna(0).astype(int)
crosstab_result['% Praticam exerccio e fumam'] = (crosstab_result['Praticam exerccio e fumam'] / crosstab_result['Pessoas que fazem exerccios']) * 100
crosstab_result['% No Praticam exerccio e fumam'] = (crosstab_result['No praticam exerccio e fumam'] / (crosstab_result['Total de pessoas'] - crosstab_result['Pessoas que fazem exerccios'])) * 100
crosstab_result.fillna(0)
```

A tabela acima apresenta novos dados em relao a anterior, onde  possvel obeservar a porcentagem de pessoas que praticam exerccios fsicos e fumam. Tambm pode ser observado e comparado a quantidade de pessoas que praticam atividades fsicas e fumam em relao as que no realizam atividades e fumam. Todos dados tambm esto divididos entre sexo.

Com esses dados em mos  possvel dizer que os homens praticantes de exerccos de forma proporcional as mulheres que tambm realizam atividades neste grupo tendem a fumar mais. Outro ponto que podemos destacar  que a porcentagem das mulheres que fumam e no realizam nenhuma atividade  maior.

```python
#Cria grfico do tipo countagem do SeaBorn
sns.countplot(data=df_filtered, x='Fuma', hue='Sexo')
plt.xlabel('Fumantes')
plt.ylabel('Quantidade de Pessoas')
plt.title('Distruio de Pessoas que Fumam')
plt.show()
```

O grfico acima  apenas um complemento da populao que foi abordado na anlise anterior.  observado que a quantidade (no proporcional) de mulheres que no fumam  maior que a de homens, porm quando  feita a comporao em porcentagem pode ser observado que a histria e diferente (conforme abordado no tpico anterior).

```python
# Criar uma tabela de frequncia cruzada
df_cinema = pd.crosstab(index=df['Turma'], columns=df['Cine'])

#Soma os valores da linha (axis=1)
df_cinema['Total'] = df_cinema.sum(axis=1)
df_cinema
```
