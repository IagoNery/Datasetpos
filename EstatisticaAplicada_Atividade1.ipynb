```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import io

from google.colab import files


uploaded = files.upload()
```

Fazer a leitura de arquivo .csv e em seguida colocar ele em um dataframe:

```python
df = pd.read_csv(io.BytesIO(uploaded['tabela1.csv']), sep=";")
df
```

Análises e Gráficos:

```python
df_sexo = pd.crosstab(index=df['Turma'], columns=df['Sexo'])
df_sexo
```

A tabela acima faz a relação entre as turmas e quantidade de alunos por sexo. É possível verificar que tanto na turma A quanto na turma o número de mulheres é maior que a de homens, isto pode ser um indicativo que o curso atrai mais o público feminino.

```python
# Filtrando apenas as linhas onde 'Exerc' é maior que 0
df_filtered = df[(df['Exerc'] > 0) & (df['Sexo'] != 'NI')]
df_filtered

# Usando a pd.crosstab para criar a tabela cruzada
crosstab_result = pd.crosstab(index=df_filtered['Sexo'], columns='Pessoas que fazem exercícios')
crosstab_result['Total de pessoas'] = df['Sexo'].value_counts()
crosstab_result ['Porcentagem (%)'] = (crosstab_result['Pessoas que fazem exercícios'] / (crosstab_result['Total de pessoas'])) * 100
crosstab_result = crosstab_result.rename_axis("", axis=1)

crosstab_result
```

Na tabela acima é feito a correlação entre sexo e a prática de exercício físico, e também é gerado a proporção desta prática em relação aos sexos. Com esses dados em mãos nota-se que os homens praticam mais exercício físico do que as mulheres de forma proporcional.

```python
#Cria filtros de dataframe para auxiliar na construção da tabela
df_filtro_fuma_exerc = df[(df['Fuma'] == 'SIM') & (df['Exerc'] > 0)]
df_filtro_fuma_nexerc = df[(df['Fuma'] == 'SIM') & (df['Exerc'] == 0)]
#df_fuma = pd.crosstab(index=df[(df['Fuma'] == 'SIM') & (df['Exerc'] > 0)]['Sexo'], columns='FumantesExercício')

# Criar uma tabela de frequência cruzada
df_fuma = pd.crosstab(index=df_filtro_fuma_exerc['Sexo'], columns='FumantesExercício')
df_fuma['FumantesSemExercício'] = df_filtro_fuma_nexerc['Sexo'].value_counts()
df_fuma['Fuma'] = df['Fuma'].value_counts()

crosstab_result['Praticam exercício e fumam'], crosstab_result['Não praticam exercício e fumam'] = df_fuma['FumantesExercício'], df_fuma['FumantesSemExercício'].fillna(0).astype(int)
crosstab_result['% Praticam exercício e fumam'] = (crosstab_result['Praticam exercício e fumam'] / crosstab_result['Pessoas que fazem exercícios']) * 100
crosstab_result['% Não Praticam exercício e fumam'] = (crosstab_result['Não praticam exercício e fumam'] / (crosstab_result['Total de pessoas'] - crosstab_result['Pessoas que fazem exercícios'])) * 100
crosstab_result.fillna(0)
```

A tabela acima apresenta novos dados em relação a anterior, onde é possível obeservar a porcentagem de pessoas que praticam exercícios físicos e fumam. Também pode ser observado e comparado a quantidade de pessoas que praticam atividades físicas e fumam em relação as que não realizam atividades e fumam. Todos dados também estão divididos entre sexo.

Com esses dados em mãos é possível dizer que os homens praticantes de exercícos de forma proporcional as mulheres que também realizam atividades neste grupo tendem a fumar mais. Outro ponto que podemos destacar é que a porcentagem das mulheres que fumam e não realizam nenhuma atividade é maior.

```python
#Cria gráfico do tipo countagem do SeaBorn
sns.countplot(data=df_filtered, x='Fuma', hue='Sexo')
plt.xlabel('Fumantes')
plt.ylabel('Quantidade de Pessoas')
plt.title('Distruição de Pessoas que Fumam')
plt.show()
```

O gráfico acima é apenas um complemento da população que foi abordado na análise anterior. É observado que a quantidade (não proporcional) de mulheres que não fumam é maior que a de homens, porém quando é feita a comporação em porcentagem pode ser observado que a história e diferente (conforme abordado no tópico anterior).

```python
# Criar uma tabela de frequência cruzada
df_cinema = pd.crosstab(index=df['Turma'], columns=df['Cine'])

#Soma os valores da linha (axis=1)
df_cinema['Total'] = df_cinema.sum(axis=1)
df_cinema
```

Outra análise simples feita aqui, foi observado que as pessoas da turma A tendem a frequentar mais ao cinema do que a turma B. Nas demais colunas é apresentado as frequências que ocorre este comparecimento ao cinemas. O gráfico abaixo reforça de forma visual estes dados e conclusão.

```python
#Cria gráfico do tipo histograma do SeaBorn
sns.histplot(data=df, x='Cine', hue='Turma')
plt.xlabel('Frequência')
plt.ylabel('Quantidade de Pessoas')
plt.title('Distruição de Pessoas Vão ao Cinema por Turma')
plt.show()
```

```python
#Cria gráfico do tipo histograma do SeaBorn
sns.histplot(data=df, x='Alt', hue='Turma', bins=20, kde=True)
plt.xlabel('Altura')
plt.ylabel('Quantidade de Pessoas')
plt.title('Distruição da Altura das Pessoa em Cada Turma')
plt.show()
```

No gráfico trás algumas informações sobre a população estudada em relação a sua altura. É observado que temos uma maior densidade (quantidade) de pessoas da turma A com altura entre as faixas 1.60 e 1.70, já turma B está entre as faixas de 1.65 a 1.75.

Outro ponto que pode se concluir ao verificar o gráfico é que a turma A em geral apresenta ter um número pessoas com estatura mais alta que a turma B.
